## Univariate and Multivariate Data Analysis

### Age Analysis

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 6))
sns.boxplot(emp_data['Age'], ax=ax[0])
sns.distplot(emp_data['Age'], ax=ax[1])
plt.show()
```

Min Age is 18 which is obivous because min age of working is 18 years
Max Age is 60 which is also the retiremet age
Most of the people are in the age between 30 - 32

### Age With Gender

```python
plt.figure(figsize=(8, 6), dpi=80)
sns.boxplot(emp_data['Gender'], emp_data['Age'])
plt.show()
```

### EducationBackground

```python
plt.figure(figsize=(12, 4), dpi=80)
sns.countplot(emp_data['EducationBackground'])
plt.show()
```

Majority of the people are from Life Sciences and then Medical background, seems like company is from Pharmaceutical domain

### Distance From Home

```python
plt.figure(figsize=(8, 6), dpi=80)
sns.distplot(emp_data['DistanceFromHome'])
plt.show()
```

Majority of People come from within 1-2km range 

### Performance Rating  

```python
plt.figure(figsize=(6, 6), dpi=80)
emp_data['PerformanceRating'].value_counts().plot('pie', autopct = '%0.02f')
plt.show()
```

- 72.83% of the people got 3 ratings
- 16.17% of the people got 2 ratings
- 11.00% of the people got 4 rating

Rating 3 seems to be the avarage rating across company so people with performance rating 2 seems to have issues

### Does age and gender impacts Performance?

```python
plt.figure(figsize=(15, 6), dpi=80)
sns.boxplot(emp_data['PerformanceRating'], emp_data['Age'], hue=emp_data['Gender'])
plt.show()
```

From the above graph, it's clearly visible that there is no relation between Age, Gender and Performance Ranting

### Does Education Background impacts Performance Ratings?

```python
plt.figure(figsize=(15, 6), dpi=80)
sns.barplot(emp_data['EducationBackground'], emp_data['PerformanceRating'])
plt.show()
```

### Does Employee Department Is related to Performace?


```python
plt.figure(figsize=(15, 6), dpi=80)
sns.barplot(emp_data['EmpDepartment'], emp_data['PerformanceRating'])
plt.show()
```

###  Is Emp Job Satisfaction impacts his/her performance?


```python
plt.figure(figsize=(12, 4), dpi=80)
(emp_data.groupby(['EmpJobSatisfaction']).size()/len(emp_data)*100).plot('bar')
plt.show()
```

Almost 19% of the people are not satisfied with their job or satisfaction is low


```python
temp = emp_data.groupby(['EmpDepartment', 'EmpJobSatisfaction']).size()
temp
```

```python
temp.groupby(level = 0).sum()
```

```python
plt.figure(figsize=(15, 4), dpi=80)
np.round(temp/temp.groupby(level=0).sum() * 100, 2)
(temp/temp.groupby(level=0).sum() * 100).plot('bar')
plt.show()
```

Employees in R&D and Sales seems to be less satisfied with the job. 
* People in these two department have approximately equal %of Low and Medium Satistfaction

### Emp Job Role

```python
# Combining Role with Department
emp_data['Department_Role'] = emp_data['EmpJobRole'] + " - " + emp_data['EmpDepartment']
emp_data['Department_Role'].unique()
temp = emp_data.groupby(['Department_Role', 'EmpJobSatisfaction']).size()
temp.groupby(level = 0).sum()

plt.figure(figsize=(20, 6), dpi=80)
np.round(temp/temp.groupby(level=0).sum() * 100, 2)
(temp/temp.groupby(level=0).sum() * 100).plot('bar')
plt.show()
```
